```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding="unicode_escape")
print(df)
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status", "unnamed1"], axis=1, inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df["Amount"] = df["Amount"].astype("int")
```

```python
df["Amount"].dtypes
```

```python
df.columns
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
sns.set({"figure.figsize": (5,5)})
ax = sns.countplot(x = "Gender", data = df)

plt.title("Customers Based on Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("Customers Based on Gender.jpg")
plt.show()
```

```python
df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by= "Amount", ascending= False)
```

```python
sns.set({"figure.figsize": (5,5)})
sales_gen = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by= "Amount", ascending= False)

plt.title("Sales Based on Gender")
sns.barplot(x="Gender", y="Amount", data= sales_gen)
plt.savefig("Sales Based on Gender.jpg")
plt.show()
```

```python
ax = sns.countplot(x = "Age Group", data = df, hue="Gender")

plt.title("Customers Based on Age Group and Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("Customers Based on Age Group and Gender.jpg")
plt.show()
```

```python
sales_age = df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by= "Amount", ascending= False)

plt.title("Sales Based on Age Group")
sns.barplot(x="Age Group", y="Amount", data= sales_age)
plt.savefig("Sales Based on Age Group.jpg")
plt.show()
```

```python
sales_state = df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by= "Orders", ascending= False).head(10)

plt.figure(figsize= (15,5))
plt.title("Orders Based on State")
sns.barplot(x="State", y="Orders", data= sales_state)
plt.savefig("Orders Based on State.jpg")
plt.show()
```

```python
sales_state = df.groupby(["State", "Gender"], as_index=False)["Amount"].sum().sort_values(by= "Amount", ascending= False).head(10)

plt.figure(figsize= (10,5))
plt.title("Sales Based on State and Gender")
sns.barplot(x="State", y="Amount", data= sales_state, hue="Gender")
plt.savefig("Sales Based on State and Gender.jpg")
plt.show()
```

```python
sns.set({"figure.figsize": (5,5)})
ax = sns.countplot(x = "Marital_Status", data = df)

plt.title("Marital Status of Customers")
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("Marital Status of Customers.jpg")
plt.show()
```

```python
sns.set({"figure.figsize": (5,5)})
sales_ms = df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by= "Amount", ascending= False)

plt.title("Sales Based on Marital Status and Gender")
sns.barplot(x="Marital_Status", y="Amount", data= sales_ms, hue="Gender")
plt.savefig("Sales Based on Marital Status and Gender.jpg")
plt.show()
```

```python
sns.set({"figure.figsize": (18,5)})
ax = sns.countplot(x = "Occupation", data = df)

plt.title("Orders Based on Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("Orders Based on Occupation.jpg")
plt.show()
```

```python
sns.set({"figure.figsize": (10,5)})
sales_occ = df.groupby(["Occupation", "Gender"], as_index=False)["Amount"].sum().sort_values(by= "Amount", ascending= False).head(10)

sns.barplot(x="Occupation", y="Amount", data= sales_occ, hue="Gender")
plt.title("Sales Based on Occupation and Gender")
plt.savefig("Sales Based on Occupation and Gender.jpg")
plt.show()
```

```python
sns.set({"figure.figsize": (25,5)})
ax = sns.countplot(x = "Product_Category", data = df)

plt.title("Orders Based on Product Category")
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("Orders Based on Product Category.jpg")
plt.show()
```

```python
sns.set({"figure.figsize": (10,5)})
sales_pc = df.groupby(["Product_Category", "Gender"], as_index=False)["Amount"].sum().sort_values(by= "Amount", ascending= False).head(10)

sns.barplot(x="Product_Category", y="Amount", data= sales_pc, hue="Gender")
plt.title("Sales Based on Product Category and Gender")
plt.savefig("Sales Based on Product Category and Gender.jpg")
plt.show()
```

```python
sns.set({"figure.figsize": (10,5)})
sales_pid = df.groupby(["Product_ID"], as_index=False)["Orders"].sum().sort_values(by= "Orders", ascending= False).head(10)

sns.barplot(x="Product_ID", y="Orders", data= sales_pid)
plt.title("Orders Based on Product ID")
plt.savefig("Orders Based on Product ID.jpg")
plt.show()
```
